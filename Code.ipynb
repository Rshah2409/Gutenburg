**Fetching a book named "Bryant-Stories" as a dataset from nltk.**


```python
#Fetching data from nltk gutenberg library
text=nltk.corpus.gutenberg.raw('bryant-stories.txt')
text
```

```python
#Length of text
print(len(text))
```

**Pre-processing the dataset.**

```python
# Dowloading, importing and printing the stopwords
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords_nltk=set(stopwords.words('english'))
stopwords_nltk
```

```python
#Converting into lower text
lower_text=text.lower()
print(lower_text)
```

```python
#Removing Punctuation

import re
text_re=re.sub(r"[-] | [_] | [--]", " ", lower_text)
text_re
```

```python
#Removing Digits
nodigits=re.sub(r"[0-9]", "", text_re)
nodigits
```

```python
#Removing Whitespace
no_whitespace=re.sub(r"\s+", " ",nodigits)
no_whitespace
```

```python
#Removing the stopwords
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('punkt')
tokens=word_tokenize(no_whitespace)

tokens_without_stopwords=[word for word in tokens if not word in stopwords.words()]
print(tokens_without_stopwords)
```

```python
#Printing Filtered Sentence
filtered_sentence=(" ").join(tokens_without_stopwords)
print(filtered_sentence)
```

**Importing DataFrames**

```python
#Importing Dataframes
from pandas import DataFrame
```

**Converting the text data into dataframe and generalizing the program.**

```python
# Initialize an empty list to store the segments
segments = []
book_id = '1918' # The Adventures of Sherlock Holmes
num_segments = 200
```

```python
# Divide the text into segments of 100 words
for i in range(num_segments):
  start = i * 100
  end = (i + 1) * 100
  segment = filtered_sentence[start:end]
  segments.append(segment)
```

```python
# Create a DataFrame with the segments
df = DataFrame(segments, columns=['text'])
```

```python
# Add a column for the book label
df['label'] = book_id
df['segment_num'] = range(1,num_segments+1)
```

```python
# Serialize the data
df.to_csv(f'{book_id}_segments.csv')
print(df)
```
